```python
folder_path = r"C:\Users\Admin\OneDrive\Documents\RAG GENERATIVE AI FULL STACK COURSE\LEC 11-12 April\data"
story=[]
for filename in os.listdir(folder_path):
  file_path=os.path.join(folder_path,filename)
  with open(file_path,encoding='unicode_escape') as f:
    corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))  # same logic as above but it will give the list of words in the sentence and it will remove the stop words and punctuations for all the files in the folder
```

```python
custom_model =gensim.models.Word2Vec(  # it will train the word2vec model on the story list which we have created above
    window=10, 
    min_count=1,
    vector_size=150
    
)
custom_model.build_vocab(story)
len(custom_model.wv.index_to_key)
```

```python
custom_model.train(story, total_examples=custom_model.corpus_count, epochs=10) # it will train the model on the story list for 10 epochs and it will give the total examples which we have in the story list
```

```python
custom_model.wv["arjun"] # it will give the vector representation of the word "arjun" which is present in the story list and it will give the vector of size 150 as we have defined the vector size as 150 in the model training above
```

```python
custom_model.wv.most_similar("arjun") # it will give the most similar words to the word "arjun" which is present in the story list and it will give the similarity score as well
```

```python
custom_model.wv.most_similar("engineer")
```

```python
custom_model.wv.similarity("ai", "technology") # it will give the similarity score between the words "ai" and "technology" which are present in the story list
```

```python
custom_model.wv.doesnt_match(["weather", "crops", "farmers", "technology"]) # it will give the word which is not similar to the other words in the list and it will give the word "technology" as it is not similar to the other words in the list as it is not related to the weather, crops and farmers
```
